```python
import pandas as pd
import numpy as np
```

```python
distric_ed_standard_rates = pd.read_csv('../data/distict_ed_standard.csv')
distric_ed_standard_rates
```

```python
distric_ed_standard_rates['district'] = distric_ed_standard_rates['DistrictName']
distric_ed_standard_rates
```

```python
distric_ed_standard_rates['district'] = distric_ed_standard_rates['DistrictName'].str.slice(stop=-16)
distric_ed_standard_rates
```

```python
distric_ed_standard_rates['district'][0] = 'Seattle'
```

```python
distric_ed_standard_rates['district'] = distric_ed_standard_rates['district'].str.upper()
distric_ed_standard_rates
```

```python
school_sites_df = pd.read_csv('../data/School_Sites_in_King_County___schsite_point.csv')
school_sites_df
```

```python
merged_df = school_sites_df.merge(distric_ed_standard_rates, left_on='DISTRICT', right_on='district', suffixes=(False, False))
```

```python
merged_df
```

```python
definitive_school_scores_df = merged_df[['ZIPCODE','percent_met_standard','district']]
definitive_school_scores_df
```
